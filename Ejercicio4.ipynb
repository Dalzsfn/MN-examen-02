```python
from scipy.optimize import bisect
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x) : 
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625
```

```python
x_lin = np.linspace(-300, 300, 400)
raices = []
for i in range(len(x_lin)-1):
    if f(x_lin[i]) * f(x_lin[i+1]) < 0:
        raiz = bisect(f, x_lin[i], x_lin[i+1])
        if round(raiz, 5) not in [round(r, 5) for r in raices]:
            raices.append(raiz)
        
print("Raíces encontradas:")
for r in raices:
    print(r)
```
